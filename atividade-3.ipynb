### 1. Escreva um programa com funções, uma para ler matrizes, outra para somá-las e uma outra para escrever seu resultado.

```python
# Função que ler valores de uma linha
def read_row(cols=3, text=''):
    row = input(text).split(' ')
    row = list(map(float, row))
    if (len(row) != cols):
        # A quantidade de valores digitados deve ser igual ao número de colunas
        raise Exception(f'A linha da matriz deve ter {cols} valores.')
    return row;

# Função para exibir linha de uma matriz
def print_row(row=[]):
    values = ' '.join(map(str, row))
    print(f'| {values} |')

# Função para somar valores de uma linha
def add_rows(row_a=[], row_b=[]):
    row_c = []
    len_a, len_b = len(row_a), len(row_b)
    if (len_a != len_b):
        # Se as linhas tiverem uma quantidade diferente de valores, uma exceção é lançada
        raise Exception(f'As linhas devem ter o mesmo número de colunas.')
    for i in range(len_a):
        row_c.append(row_a[i] + row_b[i])
    return row_c

# Função para ler uma matriz
def read_matrix(rows=3, cols=3):
    matrix = []
    for i in range(rows):
        row = read_row(cols, f'Insira os {cols} valores da linha {i + 1}:')
        matrix.append(row)
    return matrix

# Função para exibir valores de uma matriz
def print_matrix(matrix):
    for row in matrix:
        print_row(row)

# Função para somar valores de duas matrizes
def add_matrices(matrix_a, matrix_b):
    matrix_c = []
    len_a, len_b = len(matrix_a), len(matrix_b)
    if (len_a != len_b):
        # Se as matrizes tiverem uma quantidade diferente de linhas, uma exceção é lançada
        raise Exception(f'As matrizes devem ter o mesmo número de linhas.')
    for i in range(len_a):
        matrix_c.append(add_rows(matrix_a[i], matrix_b[i]))
    return matrix_c

# Função para ler as dimensões das matrizes
def read_dimensions():
    rows = input('Insira o número de linhas das matrizes:')
    rows = int(rows)
    cols = input('Insira o número de colunas das matrizes:')
    cols = int(cols)
    return (rows, cols)

# Programa
print('--------------------------------------')
print('Somar duas matrizes')
print('--------------------------------------')

# Ler número de linhas e colunas das matrizes
rows, cols = read_dimensions()

print('--------------------------------------')
print('Forneça os valores da primeira matriz:')
print('--------------------------------------')

# Ler e exibir valores da primeira matriz
matrix_a = read_matrix(rows, cols)
print_matrix(matrix_a)

print('--------------------------------------')
print('Forneça os valores da segunda matriz:')
print('--------------------------------------')

# Ler e exibir valores da segunda matriz
matrix_b = read_matrix(rows, cols)
print_matrix(matrix_b)

print('--------------------------------------')
print('Resultado da soma:')
print('--------------------------------------')

# Somar e exibir valores das matrizes
matrix_c = add_matrices(matrix_a, matrix_b)
print_matrix(matrix_c)
```

### 2. Faça um programa que leia a lista de valores, a seguir, resultada da medida de uma experiência de laboratório (peso molecular do gás e velocidade do som neste gás) e calcule, as médias aritméticas, os desvios padrões, as variâncias de cada grandeza, e a correlação entre os dados.

```python
pm = [2, 16, 17, 28, 30, 32, 44, 46, 75]
v = [1266, 431, 414, 336, 324, 316, 261, 257, 184]

# Função para calcular a média aritmética
def avg(values=[]):
    n = len(values)
    # Função built in para calcular soma
    s = sum(values)
    return (s/n)

# Função para calcular o desvio padrão
def stdev(values=[], average=None):
    n = len(values)
    if (average == None):
        average = avg(values)
    s = sum(map(lambda x: (x-average)**2, values))
    return (s/(n-1))**(1/2)

# Função para calcular o produto escalar
def dot(x=[], y=[]):
    nx, ny = len(x), len(y)
    if (nx != ny): raise Exception('As listas devem ter o mesmo tamanho.')
    res = 0
    for i in range(nx):
        res += x[i] * y[i]    
    return res

# Função para calcular os coeficientes da reta
def lin(x=[], y=[]):
    nx, ny = len(x), len(y)
    if (nx != ny): raise Exception('As listas devem ter o mesmo tamanho.')
    n = nx
    a = (dot(x,y) -1/n*sum(x)*sum(y))/(dot(x, x) -1/n*sum(x)**2)
    b = (sum(y) - a*sum(x))/n
    return (a, b)

def r_value(a, stdev_x, stdev_y):
    r = a*(stdev_x/stdev_y)
    return (r, r**2)

# Programa
avg_pm = avg(pm)
avg_v = avg(v)
stdev_pm = stdev(pm, avg_pm)
stdev_v = stdev(v, avg_v)
a, b = lin(pm, v)
r, r2 = r_value(a, stdev_pm, stdev_v)

print(f'Média dos pesos moleculares (PM):        {avg_pm}')
print(f'Média dos valores de velocidade (V):     {avg_v}')
print(f'Desvio padrão dos pesos moleculares:     {stdev_pm}')
print(f'Desvio padrão dos valores de velocidade: {stdev_v}')
print(f'Equação da reta (PM x Velocidade):       {a}*x + {b}')
print(f'Correlação entre os valores (R):         {r}')
print(f'Correlação entre os valores (R²):        {r2}')
```
